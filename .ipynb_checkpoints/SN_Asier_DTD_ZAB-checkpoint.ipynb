```python
#!/usr/bin/python

import numpy as np
import math
from astropy.io import fits as fits

from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm

import pandas as pd
from pandas import DataFrame, Series

from astropy.wcs import WCS as WCS

import import_ipynb
import Pipe3D_Asier_Test_ZAB as Pipe3D

from mpdaf.obj import hms2deg
from mpdaf.obj import dms2deg

from astropy.convolution import Gaussian2DKernel
from astropy.convolution import convolve

from scipy.stats import ks_2samp
from scipy import stats
```

```python
alfa = 1.0
delta = 0.0
sigma = 20
```

```python
#Crear arrays con los datos de todas las galaxias.
#Numb_Lines = len(open('Names_Amusing_true.txt').readlines())
Numb_Lines = len(open('Names_Amusing_2.5.txt').readlines())

All_SN_type = np.zeros(Numb_Lines)
All_Prob_ZaB = np.zeros(Numb_Lines)
All_Cumul_ZaB = np.zeros(Numb_Lines)

#names = open('Names_Amusing_true.txt')
names = open('Names_Amusing_2.5.txt')

Galaxy = -1
nIa = 0
nII = 0
for lines in names:
    line= lines[:-1]
    print(line)
    Galaxy = Galaxy+1

    #df = pd.read_csv('Amusing_Table_true.csv')
    df = pd.read_csv('Amusing_Table_2.0.csv')  # To find the Galaxy data
    Names_1 = df['SN_name']
    Names_2 = df['def_name']
    R = 0

    for names in Names_1:              # Finding the Galaxy in the DataFrame
        R=R+1
        if line == names:
            Row = R-1
    R = 0
    for names in Names_2:              # Some Galaxies have the name in other column
        R=R+1
        if line == names:
            Row = R-1

    # SN coordinates
    RA_SN = df.at[Row,'RA_SN']
    RA_SN = hms2deg(RA_SN)

    DEC_SN = df.at[Row,'DEC_SN']
    DEC_SN = dms2deg(DEC_SN)

    # Galaxy center coordinates
    RA_GAL = df.at[Row,'RA_GAL']
    RA_GAL = hms2deg(RA_GAL)

    DEC_GAL = df.at[Row,'DEC_GAL']
    DEC_GAL = dms2deg(DEC_GAL)

    # Pixel position
    #w = WCS('../Data/Amusing/flux_elines.'+line+'.cube.fits')
    w = WCS('../Data2.0/flux_elines.'+line+'.cube.fits')
    Pix_X, Pix_Y ,z = w.wcs_world2pix(RA_SN, DEC_SN, 0. , 0,ra_dec_order=True)
    Pix_X_GAL, Pix_Y_GAL ,z = w.wcs_world2pix(RA_GAL, DEC_GAL, 0. , 0,ra_dec_order=True)

    # Integer for measure
    Pix_X = int(Pix_X)
    Pix_Y = int(Pix_Y)
    
    # Supernova type
    if df.at[Row,'SN_type'] == 'Ia':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'Iax':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'Ia-00cx':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'Ia-pec':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'Ia-91T':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'Ia-p':
        SN_TYPE = 1
    elif df.at[Row,'SN_type'] == 'II':
        SN_TYPE = 2
    elif df.at[Row,'SN_type'] == 'IIP':
        SN_TYPE = 2
    elif df.at[Row,'SN_type'] == 'IIb':
        SN_TYPE = 2
    elif df.at[Row,'SN_type'] == 'IIn':
        SN_TYPE = 2
    elif df.at[Row,'SN_type'] == 'Ib':
        SN_TYPE = 3
    elif df.at[Row,'SN_type'] == 'Ibn':
        SN_TYPE = 3
    elif df.at[Row,'SN_type'] == 'Ibc':
        SN_TYPE = 3
    elif df.at[Row,'SN_type'] == 'Ic':
        SN_TYPE = 3
    elif df.at[Row,'SN_type'] == 'Ic-BL':
        SN_TYPE = 3
    else:
        SN_TYPE = 0

#~~~~~~~~~~~~ SN Rates ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

    # Call to Pipe3D code
    line = Pipe3D.Pipe3D_galaxy(line)

    # Obtain the maps of the galaxy
    line.painting_galaxy_maps(False, alfa, delta, sigma)

    # Gaussian function
    stddev = 7
    gauss = Gaussian2DKernel(stddev)

    # Gaussian convolve
    line.SNZaB_rate_map = convolve(line.SNZaB_rate_map, gauss)

    # Eliminate negative values
    line.SNZaB_rate_map[line.SNZaB_rate_map<0.0] = 0.0

# ------------- Mass ------------------------------------------------------------------------------------

    # An Array whit all the value of the SN Ia rate map
    line.Prob_List_ZaB = np.ndarray.flatten(line.SNZaB_rate_map)
    # The array orther in value
    line.Order_Prob_List_ZaB = np.sort(line.Prob_List_ZaB)

    # Normalize
    line.Norm_SNZaB = 0
    for i in range(len(line.Order_Prob_List_ZaB)):
        line.Norm_SNZaB = line.Norm_SNZaB + line.Order_Prob_List_ZaB[i]

    line.Order_Prob_List_ZaB = line.Order_Prob_List_ZaB / line.Norm_SNZaB
    line.SNZaB_rate_map = line.SNZaB_rate_map / line.Norm_SNZaB

    # Calculate P
    line.cumulative_P_values_ZaB = np.cumsum(line.Order_Prob_List_ZaB)

    # Calculate P map
    line.cumulative_SNZaB_rate_map = np.interp(line.SNZaB_rate_map, line.Order_Prob_List_ZaB, line.cumulative_P_values_ZaB)

    # Measure Probability
    line.Prob_Val_SNZaB = 0
    line.Cumulative_Prob_SNZaB = 0
    square = np.arange(3)-1

    for j in square:
        for i in square:
            # The probability of SNIa (p)
            line.Prob_Val_SNZaB = line.Prob_Val_SNZaB + line.SNZaB_rate_map[Pix_Y+j,Pix_X+i]
            # The cumulative probability of SNIa (P)
            line.Cumulative_Prob_SNZaB = line.Cumulative_Prob_SNZaB + line.cumulative_SNZaB_rate_map[Pix_Y+j,Pix_X+i]

    line.Prob_Val_SNZaB = line.Prob_Val_SNZaB / 9
    line.Cumulative_Prob_SNZaB = line.Cumulative_Prob_SNZaB/9
    
    # ------------- Save Data ------------------------------------------------------------------------------------

    # SN type
    All_SN_type[Galaxy] = SN_TYPE
    # SN Ia and cumulative SN Ia Probability
    All_Prob_ZaB[Galaxy] = line.Prob_Val_SNZaB
    All_Cumul_ZaB[Galaxy] = line.Cumulative_Prob_SNZaB

    data_ZaB = np.column_stack((All_SN_type,All_Prob_ZaB,All_Cumul_ZaB))
    np.savetxt('../Results_Zb.txt',data_ZaB)
```
